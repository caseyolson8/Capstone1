```python
a,b = params
Y_mod = X[:,0]*a + b
n = len(Y)
t_stat = stats.t.ppf(0.975, n - 2)
sigma = np.sqrt(((Y_mod - Y)**2).sum()/(len(Y)-2))

den = sum((X[:,0]-X[:,0].mean())**2)


print(sigma)
print(sigma**2)
```
